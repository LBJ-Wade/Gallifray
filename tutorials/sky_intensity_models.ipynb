## This tutorial describes the basic usuage of `Gallifray` models for producing sky images with the `gallifray.models` module

```python
import numpy as np
import gallifray as gr
import matplotlib.pyplot as plt
```

## Setting up parameters for the model and initialising the image object

```python
I0 = 1
Rp = 40
Rn = 29
ecn = -0.42
f = 0.1
phi = 17*np.pi/180
gax=0.2
aq=1
gq=0
fov = Rp*2 + 20
xs1 = gr.models.xsringauss(I0,Rp,Rn,ecn,f,gax=gax,aq=aq,gq=gq,phi=phi,fov=fov)
```

```python
im = xs1.sky_map() # Returns the 2-D array image
plt.imshow(im, cmap='afmhot')
```

## Convolving the image with the Interstellar scattering kernel (default sets for Sag A*)

```python
im1 = xs1.sky_blur() # Returns the 2-D array image
plt.imshow(im1, cmap='afmhot')
```
